```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def analyze_df_stats(df):
    # Calculate new columns based on provided formulas
    df['Sample Pos Rate in %'] = np.round((df['Positive Length'] / df['Total Length']) * 100, 2)
    df['Sample Pos Diversity Rate in %'] = np.round((df['Unique CC Count'] / df['Tumor idx']) * 100, 2)
    df['Sample Pos Pixel Rate in %'] = np.round((df['Pos Pixel Count'] / df['Total Size']) * 100, 2)
    
    # Outputting sum statistics
    pos_length_sum = df['Positive Length'].sum()
    neg_length_sum = df['Negative Length'].sum()
    unique_cc_count_sum_ratio = df['Unique CC Count'].sum() / df['Tumor idx'].sum() * 100
    pos_pixel_count_sum_ratio = df['Pos Pixel Count'].sum() / df['Total Size'].sum() * 100
    total_number_tumors = df['Tumor idx'].sum()
    total_number_tumors_identified = df['Unique CC Count'].sum()
    total_number_tumors_missed = total_number_tumors - total_number_tumors_identified
    
    print(f"Total Positive Length: {pos_length_sum}")
    print(f"Total Negative Length: {neg_length_sum}")
    print(f"Average Unique CC Count / Tumor idx Ratio: {unique_cc_count_sum_ratio}")
    print(f"Average Pos Pixel Count / Total Size Ratio in %: {pos_pixel_count_sum_ratio}")
    print(f'Total Number of Tumors: {total_number_tumors}')
    print(f'Total Number of Tumors Identified: {total_number_tumors_identified}')
    print(f'Total Number of Tumors Missed: {total_number_tumors_missed}')


    # Visualization
    plt.figure(figsize=(18, 6))

    plt.subplot(1, 3, 1)
    plt.hist(df['Sample Pos Rate in %'], color='skyblue')
    plt.title('Sample Positive Rate Distribution', fontsize=20)
    plt.xlabel('Sample Positive Rate (%)', fontsize=15)
    plt.ylabel('Number of Samples', fontsize=15)

    plt.subplot(1, 3, 2)
    plt.hist(df['Sample Pos Diversity Rate in %'], color='lightgreen')
    plt.title('Sample Positive Diversity Rate Distribution', fontsize=20)
    plt.xlabel('Sample Positive Diversity Rate (%)', fontsize=15)
    plt.ylabel('Number of Samples', fontsize=15)

    plt.subplot(1, 3, 3)
    plt.hist(df['Sample Pos Pixel Rate in %'], color='salmon')
    plt.title('Sample Positive Pixel Rate Distribution', fontsize=20)
    plt.xlabel('Sample Positive Pixel Rate (%)', fontsize=15)
    plt.ylabel('Number of Samples', fontsize=15)

    plt.tight_layout()
    plt.show()

# Example usage:
# Assuming `df` is your DataFrame:
# analyze_df_stats(df)
```

```python
import numpy as np

# Correcting the error by defining np and plotting the bar graphs again.

# Function to plot a grouped bar chart
def plot_grouped_bar(data, title, y_label, log_scale=False):
    # Set up the matplotlib figure
    fig, ax = plt.subplots(figsize=(14, 8))
    
    # Number of groups
    n_groups = len(data)
    
    # Create an index for each tick position
    index = np.arange(n_groups)
    
    # The bar width
    bar_width = 0.15
    
    # The x position of bars
    positions = [index + bar_width*i for i in range(len(data.columns))]

    for pos, column in zip(positions, data.columns):
        plt.bar(pos, data[column], bar_width, label=column)
    
    # Set the y scale to logarithmic if specified
    if log_scale:
        ax.set_yscale('log')
    
    # Insert a legend outside of the plot area
    ax.legend(title='Parameter Sets', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=15)
    
    # Set labels and title
    ax.set_xlabel('Data Type', fontsize=15)
    ax.set_ylabel(y_label, fontsize=15)
    ax.set_title(title, fontsize=20)
    
    # Set the position of the x ticks
    ax.set_xticks(index + bar_width / 2 * (len(data.columns) - 1))
    ax.set_xticklabels(data.index)
    
    # Adding some aesthetics
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    
    # Rotate the tick labels for visibility
    plt.xticks(rotation=45, fontsize=15)
    plt.yticks(rotation=45, fontsize=15)

    
    # Show the plot
    plt.tight_layout()
    plt.show()

# Given data from the image provided by the user
data = {
    "Param Set": ["R=10, TH=2", "R=3, TH=2", "R=1, TH=2", "R=1, TH=1"],
    "Total # of Candidates": [169818, 2025792, 6696111, 19137733],
    "# of Neg Cand.": [164733, 206714, 664656, 19087193],
    "# of Pos Cand.": [5085, 19078, 49547, 50540],
    "Pos/Total": [2.99, 0.97, 0.74, 0.26],
    "Total # of Tumors": [7574, 7574, 7574, 7574],
    "# of Tumors Missed": [3702, 1231, 718, 672],
    "# of Tumors Identified": [3872, 6343, 6856, 6902],
    "Rate of Tumors Identified": [51.12, 83.75, 90.52, 91.13],
    "Covered Rate in Voxel": [1.48, 5.30, 12.40, 12.46]
}

# Convert to DataFrame for easier handling
df = pd.DataFrame(data)


count_columns = ["Total # of Candidates", "# of Neg Cand.", "# of Pos Cand.", "Total # of Tumors", "# of Tumors Missed", "# of Tumors Identified"]
# Creating the DataFrame for count data and plotting
count_data = df.set_index('Param Set')[count_columns].T
plot_grouped_bar(count_data, 'Count Data Grouped by Parameter Sets', 'Counts', log_scale=True)

# Creating the DataFrame for percentage data and plotting
percent_columns = ["Pos/Total", "Rate of Tumors Identified", "Covered Rate in Voxel"]
percent_data = df.set_index('Param Set')[percent_columns].T
plot_grouped_bar(percent_data, 'Percentage Data Grouped by Parameter Sets', 'Percentage')
```
